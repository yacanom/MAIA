## Preprocesamiento y Vectorizacin del Texto

El preprocesamiento para el modelado de tpicos implica transformar el texto sin estructurar en una representacin numrica que los algoritmos puedan procesar. Para este fin, utilizaremos **CountVectorizer** de **Scikit-learn**. Este proceso convierte cada documento en un vector que cuenta la frecuencia de cada palabra. Es crucial eliminar las 'stop words' (palabras comunes como 'el', 'la', 'un') para que el modelo se enfoque en palabras ms significativas para la identificacin de tpicos.

El resultado ser una Matriz Documento-Trmino (DTM), donde las filas representan documentos y las columnas representan palabras del vocabulario. Tambin se mostrar el vocabulario resultante.

```python
url = 'https://raw.githubusercontent.com/yacanom/MAIA/refs/heads/main/df_total.csv'

try:
    df = pd.read_csv(url)
    print(f"Dataset cargado correctamente con {df.shape[0]} filas y {df.shape[1]} columnas.")
except Exception as e:
    print(f"Error al cargar el dataset: {e}")

df.head()
```

El nico paso de preprocesamiento necesario en nuestro caso es la construccin de caractersticas, en la que tomamos el texto de las noticias y los representamos en algn espacio de caractersticas manejable. En la prctica, esto significa simplemente convertir cada cadena en un vector numrico.

```python
# Vectorizacion con stopwords en Spanish
spanish_stop_words = stopwords.words('spanish')

def preprocess(text):
    # Minsculas
    text = text.lower()
    # Tokenizacin simple
    tokens = text.split()
    # Eliminar stopwords
    tokens = [t for t in tokens if t not in spanish_stop_words]
    return " ".join(tokens)

data['clean_news'] = df['news'].astype(str).apply(preprocess)
data[['news','clean_news']].head()
```

```python
vectorizer = CountVectorizer(stop_words=spanish_stop_words, max_features=40000)
document_term_matrix = vectorizer.fit_transform(data['clean_news'])

# Obtener el vocabulario
vocabulary = vectorizer.get_feature_names_out()

# Ejemplo: Obtener el vector de un texto especfico
text_to_vectorize = data['news'].iloc[0]  # Obtener el primer texto
text_vector = vectorizer.transform([text_to_vectorize]).toarray()

print('Vector del texto:', text_vector)
print('Vocabulario:', vocabulary)
```

# **Modelamiento**

```python
# Se define un scorer no supervisado basado en error de reconstruccin.
def lsa_reconstruction_scorer(estimator, X, y=None):
    # Se extraen pasos del pipeline
    vec = estimator.named_steps['vectorizer']
    lsa = estimator.named_steps['lsa']
    # Se vectoriza el conjunto de validacin
    X_dtm = vec.transform(X)
    # Se proyecta y reconstruye
    X_proj = lsa.transform(X_dtm)
    X_recon = lsa.inverse_transform(X_proj)
    # Se calcula MSE de reconstruccin (menor = mejor).
    # Se devuelve NEGADO para que GridSearchCV maximice el score.
    return -mean_squared_error(X_dtm.toarray(), X_recon)

recon_scorer = make_scorer(lsa_reconstruction_scorer, greater_is_better=True)

# Pipeline: vectorizacin + LSA
lsa_pipeline = Pipeline([
    ('vectorizer', CountVectorizer(stop_words=spanish_stop_words,
                                   max_features=40000,   # ajusta si hay memoria justa
                                   ngram_range=(1,1))), # unigramas tpicos para LSA bsico
    ('lsa', TruncatedSVD(random_state=42, algorithm='randomized'))
])

# Rango de parmetros a explorar
lsa_param_grid = {
    'lsa__n_components': [3, 4, 5, 6, 7, 8, 10, 15]
}

lsa_search = GridSearchCV(
    lsa_pipeline,
    param_grid=lsa_param_grid,
    cv=3,
    scoring=recon_scorer,   # <- scorer no supervisado
    n_jobs=-1,
    verbose=2
)

lsa_search.fit(data['news'])
print("Mejores parmetros LSA:", lsa_search.best_params_)
print("Mejor (MSE) de reconstruccin (cv):", lsa_search.best_score_)
```

#### Bsqueda de hiperparmetros

Optimizar los hiperparmetros de LDA es crucial para obtener tpicos coherentes y bien definidos. Exploramos diferentes configuraciones para el nmero de tpicos (n_components), el mtodo de aprendizaje (learning_method) y el decaimiento del aprendizaje (learning_decay).

Para LDA, las mtricas ms robustas y comnmente aceptadas son la perplejidad (a minimizar) y la coherencia de tpicos (C_v) (a maximizar). La perplejidad mide qu tan bien el modelo predice el conjunto de datos (menor es mejor), mientras que la coherencia C_v evala la interpretabilidad de los tpicos al medir la similitud semntica de sus palabras ms prominentes (mayor es mejor).

Tambin se puede utilizar el log-likelihood (a maximizar), que indica la probabilidad de que el modelo haya generado el corpus observado. Sin embargo, la coherencia C_v a menudo se correlaciona mejor con la evaluacin humana de la calidad de los tpicos.

##### 1. Optimizacin de LDA usando Perplexity

La perplejidad es una mtrica estadstica que mide cun bien un modelo probabilstico predice un conjunto de datos, siendo un valor menor, mejor.

```python
# Rango de hiperparmetros a probar
param_grid = {
    "n_components": [3, 4, 5, 6, 7, 8, 9, 10],  # nmero de tpicos
    "learning_decay": [0.5, 0.7, 0.9],         # factor de aprendizaje
    "learning_method": ["online", "batch"]     # mtodo de entrenamiento
}

best_perplexity = float("inf")
best_model_perplexity = None
best_params_perplexity = None

# Bucle de bsqueda
for n in param_grid["n_components"]:
    for decay in param_grid["learning_decay"]:
        for method in param_grid["learning_method"]:
            lda = LatentDirichletAllocation(
                n_components=n,
                learning_decay=decay,
                learning_method=method,
                random_state=0 # Se fija random_state para reproducibilidad [22-24]
            )
            lda.fit(document_term_matrix)

            perplexity = lda.perplexity(document_term_matrix)
            print(f"n={n}, decay={decay}, method={method}  perplexity={perplexity:.2f}")

            if perplexity < best_perplexity:
                best_perplexity = perplexity
                best_model_perplexity = lda
                best_params_perplexity = {"n_components": n, "learning_decay": decay, "learning_method": method}

print("\nMejores parmetros LDA (Perplejidad):", best_params_perplexity)
print("Mejor perplexity:", best_perplexity)
```

En este caso el nmero de tpicos es de 6.

##### 2. Optimizacin de LDA usando Coherencia de Tpicos (C_v)

La coherencia de tpicos (C_v) evala la interpretabilidad de los tpicos midiendo la similitud semntica de sus palabras ms prominentes, siendo un valor mayor, mejor. Este enfoque utiliza la librera gensim.

```python
#Prepara los textos como listas de palabras ---
texts = [doc.split() for doc in data['news']]

#Diccionario y corpus en formato gensim ---
dictionary = corpora.Dictionary(texts)
corpus = [dictionary.doc2bow(text) for text in texts]

#Definir rango de hiperparmetros a evaluar
param_grid_coherence = {
    "num_topics": [3, 4, 5, 6, 7, 8, 9, 10],  # nmero de tpicos
    "passes": [5, 10, 19],    # nmero de pasadas sobre el corpus
    "alpha": ['symmetric', 'asymmetric'], # priors de distribucin
}

best_coherence = -1
best_model_coherence = None
best_params_coherence = None

# --- 4. Bucle de bsqueda --- [27]
for n in param_grid_coherence["num_topics"]:
    for p in param_grid_coherence["passes"]:
        for a in param_grid_coherence["alpha"]:
            lda_model = gensim.models.LdaModel(
                corpus=corpus,
                id2word=dictionary,
                num_topics=n,
                passes=p,
                alpha=a,
                random_state=0, # Se fija random_state para reproducibilidad
                eval_every=None # Se desactiva evaluacin interna
            )
            coherence_model = CoherenceModel(
                model=lda_model,
                texts=texts,
                dictionary=dictionary,
                coherence='c_v'
            )
            coherence_score = coherence_model.get_coherence()

            print(f"n={n}, passes={p}, alpha={a}  coherence={coherence_score:.4f}")

            if coherence_score > best_coherence:
                best_coherence = coherence_score
                best_model_coherence = lda_model
                best_params_coherence = {"num_topics": n, "passes": p, "alpha": a}

print("\nMejores parmetros (Coherencia C_v):", best_params_coherence)
print("Mejor coherencia (C_v):", best_coherence)
```

En este caso el nmero de tpicos es de 4.

##### Optimizacin de LDA usando Log-likelihood con GridSearchCV

El log-likelihood (a maximizar) indica la probabilidad de que el modelo haya generado el corpus observado. Este mtodo utiliza GridSearchCV de sklearn con un scorer personalizado.

```python
# Se asegura la lista de textos de entrada; se asume que 'data' posee la columna 'news'.
texts = data['news'].astype(str).tolist()

# Se definen las stopwords en espaol; si ya existen en memoria, se reutilizan.
try:
    spanish_stop_words = spanish_stop_words
except NameError:
    from nltk.corpus import stopwords
    spanish_stop_words = stopwords.words('spanish')

# Se construye el pipeline Vectorizador (BoW) -> LDA con parmetros base reproducibles.
lda_pipeline = Pipeline([
    ('vectorizer', CountVectorizer(
        stop_words=spanish_stop_words,   # Se eliminan palabras vacas en espaol.
        max_features=40000,              # Se limita el vocabulario por memoria/ruido.
        ngram_range=(1,1)                # Se emplean unigramas (recomendado para LDA clsico).
    )),
    ('lda', LatentDirichletAllocation(
        random_state=42,                 # Se fija semilla para reproducibilidad.
        max_iter=20,                     # Se fija nmero de iteraciones EM.
        evaluate_every=0,                # Se desactiva evaluacin interna (la hace GridSearch).
        n_jobs=-1                        # Se paraleliza cuando es posible.
    ))
])

# Se define un scorer no supervisado que maximiza el log-likelihood de LDA sobre X_val.
def lda_loglik_scorer(estimator, X, y=None):
    # Se obtiene el vectorizador y el modelo ya ajustados en el fold de entrenamiento.
    vec = estimator.named_steps['vectorizer']
    lda = estimator.named_steps['lda']
    # Se transforma el conjunto de validacin a matriz documentotrmino (DTM).
    X_dtm = vec.transform(X)
    # Se retorna el log-likelihood del modelo sobre la DTM (valor mayor es mejor).
    return lda.score(X_dtm)

# Se envuelve el scorer para que GridSearch lo maximice.
loglik_scorer = make_scorer(lda_loglik_scorer, greater_is_better=True)

# Se define la grilla de hiperparmetros a explorar.
# Nota: 'learning_decay' solo aplica a 'online'; para 'batch' se ignora sin afectar el scoring.
lda_param_grid = {
    'lda__n_components': [3, 4, 6, 8, 10, 15, 20],  # Se exploran tamaos de K.
    'lda__learning_method': ['online', 'batch'],    # Se compara 'online' vs 'batch'.
    'lda__learning_decay': [0.5, 0.7, 0.9]          # Se prueba decaimiento (si 'online').
}

# (Opcional) Se realiza muestreo si el corpus es grande para acelerar la bsqueda.
rng = np.random.RandomState(42)
n = len(texts)
frac = 1.0 if n <= 8000 else 0.3          # Se usa 30% para corpus grandes.
subset_idx = rng.choice(n, size=int(n*frac), replace=False)
texts_sample = [texts[i] for i in subset_idx]

# Se configura la bsqueda con validacin cruzada y el scorer no supervisado.
lda_search = GridSearchCV(
    lda_pipeline,
    param_grid=lda_param_grid,
    cv=3,
    scoring=loglik_scorer,   # <- Se corrige la mtrica: log-likelihood en lugar de MSE.
    n_jobs=-1,
    verbose=2
)

# Se ajusta la bsqueda sobre el subconjunto (o el total si n es pequeo).
lda_search.fit(texts_sample)

# Se imprimen los mejores hiperparmetros y el mejor score de validacin cruzada.
print("Mejores parmetros LDA:", lda_search.best_params_)
print("Mejor log-likelihood (cv):", lda_search.best_score_)

# Se reentrena el mejor pipeline sobre TODO el corpus para producir el modelo final.
best_lda_pipeline = lda_search.best_estimator_
best_lda_pipeline.fit(texts)

# Se derivan objetos tiles para anlisis posterior (DTM, distribucin doc-tpico, vocabulario, etc.).
best_vec = best_lda_pipeline.named_steps['vectorizer']
best_lda = best_lda_pipeline.named_steps['lda']
DTM = best_vec.transform(texts)                # Se obtiene la DTM final.
THETA = best_lda.transform(DTM)                # Se calcula distribucin de tpicos por documento.
PHI = best_lda.components_                     # Se extraen pesos palabra por tpico.
vocab = np.array(best_vec.get_feature_names_out())

# (Opcional) Se listan las top palabras por tpico para inspeccin.
def top_words_per_topic(phi, vocab, topn=10):
    tops = []
    for k in range(phi.shape[0]):
        top_idx = np.argsort(phi[k])[::-1][:topn]
        tops.append(vocab[top_idx].tolist())
    return tops

top10 = top_words_per_topic(PHI, vocab, topn=10)
for k, words in enumerate(top10, 1):
    print(f"Tpico {k:02d}: {', '.join(words)}")
```
